```python
import pandas as pd

df_train = pd.read_csv('data/train.csv')
df_val = pd.read_csv('data/val.csv')
df_test = pd.read_csv('data/test.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

import math

sns.set()

def explore(df):
    plt.figure(figsize=(15, 10))
    # class distribution
    sns.countplot(x='label', data=df)
    plt.title(f'Class Distribution')
    plt.show()
    
    # subject distribution
    plt.figure(figsize=(15, 10))
    sns.countplot(x='subject', data=df)
    plt.title(f'Subject Distribution')
    plt.show()
    
    # class distribution for each subject
    plt.figure(figsize=(32, 64))
    sns.countplot(y='subject', data=df, hue='label')
    plt.title(f'class Distribution with Respect to The Subject')
    plt.show()
    
    plt.figure(figsize=(32, 64))
    cols = math.ceil(df.subject.nunique() / 4)
    for idx, subject in enumerate(df.subject.unique()):
        plt.subplot(cols, 4, idx+1)
        sns.countplot(x='label', data=df[df.subject == subject])
        plt.title(f'Class Distribution of Subject {subject}')
    plt.show()
```

```python
explore(df_train)
```

```python
explore(df_val)
```

```python
explore(df_test)
```
